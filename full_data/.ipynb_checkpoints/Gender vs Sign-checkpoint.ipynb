```python
print("INSTAGRAM")
print()
data2=data.loc[data.platform=="instagram"]
data3=data2.loc[0<data.positive_words]
data4=data3.loc[data.positive_words>=0]
data5=data4.loc[data.male_words>=0]
data6=data5.loc[data.positive_words>=0]
print(str(len(data2.loc[0<data.positive_words])/len(data2))+" of the posts have positive words")
print(str(len(data2.loc[0<data.male_words])/len(data2))+" of the posts have male words")
print(str(len(data2.loc[0<data.negative_words])/len(data2))+" of the posts have negative words")
print(str(len(data2.loc[0<data.female_words])/len(data2))+" of the posts have female words")
print()
print(str(len(data2.loc[data.male_words<data.positive_words])/len(data2))+" of the posts have more positive words than male words")
print(str(len(data2.loc[data.male_words<data.female_words])/len(data2))+" of the posts have more female words than male words")
print(str(len(data2.loc[data.female_words<data.male_words])/len(data2))+" of the posts have more male words than female words")
print(str(len(data2.loc[data.female_words<data.negative_words])/len(data2))+" of the posts have more negative words than female words")


print()
print(str(len(data2.loc[15<data.word_count])/len(data2))+" of the posts have more than 15 words")





del data2
del data3
del data4
del data5
del data6
```
