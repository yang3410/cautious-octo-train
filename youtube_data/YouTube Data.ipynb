```python
ytvideos = pd.read_csv("100randomytvideos.csv")
ytvideos
```

Now I use the comment scrapper to produce 100 data files with comments on them: for each video, I will extract 300 random comments from them, or all of them in case they have less than 300.

```python
from itertools import islice
from youtube_comment_downloader import *
downloader = YoutubeCommentDownloader()
```

```python
for i in range(100):
    comments = downloader.get_comments_from_url(ytvideos.Video[i])
    for comment in islice(comments, 350):
        ytdata.loc[len(ytdata)] = [comment["text"]]
```

```python
ytdata
```

Let us now clean the data again:

```python
ytdata = ytdata.dropna()
ytdata = ytdata.drop_duplicates()
ytdata.reset_index(drop=True, inplace=True)
ytdata
```

Finally, we create the data file.

```python
ytdata.to_csv('ytdata.csv', index=False)
```
