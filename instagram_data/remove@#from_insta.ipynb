```python
df = pd.read_csv('insta.csv')
```

```python
df
```

```python
import re

def removingat(string):
    return re.sub('@[\w]+','',string)

def removingdash(string):
    return re.sub('#[\w]+','',string)
```

```python
df['Caption'] = df['Caption'].apply(removingat)
```

```python
df['Caption'] = df['Caption'].apply(removingdash)
```

```python
df
```

```python
df.to_csv('insta_clean.csv', index=False)
```
