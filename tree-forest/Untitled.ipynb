```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('../max72k_data/max72kpost_data.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
df = df[["platform","positive_words", "negative_words", "length", "word_count", "male_words", "female_words", "afinn", "intensity", "positivity"]]
```

```python
df = df.loc[df["word_count"]>0]
```

```python
df.shape
```

```python
df["positive_density"] = df["positive_words"]/df["word_count"]
```

```python
df["negative_density"] = df["negative_words"]/df["word_count"]
```

```python
df["male_density"] = df["male_words"]/df["word_count"]
```

```python
df["female_density"] = df["female_words"]/df["word_count"]
```

```python
sns.pairplot(df[["positive_density","negative_density","male_density","female_density","platform"]].sample(100),
             hue="platform")
```
